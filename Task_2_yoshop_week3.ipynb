```python
#create a PDF file object
pdf = PdfPages('myplots.pdf')

def analyze_reviews():
  fig=plt.figure()
  stars_index = review['stars'].value_counts().keys()[:5]
  stars_values = review['stars'].value_counts()[:5]
  
  fig, ax = plt.subplots()

  # Plot the distribution of review scores with added space between bars
  ax.bar(stars_index, stars_values, width=0.5)
  ax.set_title('Distribution of Review Scores')
  ax.set_xlabel('Stars', rotation=5)
  ax.set_ylabel('freq')
  pdf.savefig(fig) #save the current figure into a PDF page

def analyze_payment():
  fig=plt.figure()
  payment_index=order_20['Payment Method'].value_counts().keys()[:3]
  payment_values=order_20['Payment Method'].value_counts()[:3]
  fig, ax=plt.subplots()
  ax.bar(payment_index, payment_values, width=0.5)
  ax.set_title('analysis of different payment methods used by the Customers')
  ax.set_xlabel('Payment_Method')
  ax.set_ylabel('freq')
  pdf.savefig(fig) #save the current figure into a PDF page

def analyze_state():
  fig=plt.figure()
  state_index=order_20['Shipping State'].value_counts().keys()[:5]
  state_values=order_20['Shipping State'].value_counts()[:5]
  fig, ax=plt.subplots()
  ax.bar(state_index, state_values, width=0.5)
  ax.set_title('analysis of Top Consumer States of India')
  ax.set_xlabel('Consumer state')
  ax.set_ylabel('freq')
  pdf.savefig(fig) #save the current figure into a PDF page

def analyze_city():
  fig=plt.figure()
  state_index=order_20['Shipping City'].value_counts().keys()[:5]
  state_values=order_20['Shipping City'].value_counts()[:5]
  fig, ax=plt.subplots()
  ax.bar(state_index, state_values, width=0.5)
  ax.set_title('analysis of Top Consumer City of India')
  ax.set_xlabel('Consumer City')
  ax.set_ylabel('freq')

def analyze_products():
  fig=plt.figure()
  state_index=order_20['LineItem Name'].value_counts().keys()[:2]
  state_values=order_20['LineItem Name'].value_counts()[:2]
  fig, ax=plt.subplots()
  ax.bar(state_index, state_values, width=0.5)
  ax.set_title('analysis of Top Selling Product Categories')
  ax.set_xlabel('Product Categories')
  ax.set_ylabel('freq')
  plt.show()
  pdf.savefig(fig) #save the current figure into a PDF page
  
def analyze_r_category():
  fig=plt.figure()
  reviewed_category = review.groupby('category')['stars'].count()
  reviewed_category.plot.bar(figsize=(10,6))
  plt.title(' analysis of Reviews for All Product Categories')
  plt.xlabel('ALL product Category')
  plt.ylabel('Reviews')
  plt.show()
  pdf.savefig(fig) #save the current figure into a PDF page

def analyze_No_order_YM():
  fig=plt.figure()
  order_20['Order Date and Time Stamp'] = pd.to_datetime(order_20['Order Date and Time Stamp'])
  order_20['year'] = order_20['Order Date and Time Stamp'].dt.year
  order_20['month'] = order_20['Order Date and Time Stamp'].dt.month
  orders_by_month_year = order_20.groupby(['year', 'month'])['Order #'].count()
  orders_by_month_year.plot.bar(figsize=(10,6))
  plt.title('Number of Orders Per Month Per Year')
  plt.xlabel('Year, Month')
  plt.ylabel('Number of Orders')
  plt.show()
  pdf.savefig(fig) #save the current figure into a PDF page

def analyze_r_No_order_YM():
  fig=plt.figure()
  order_20['Order Date and Time Stamp'] = pd.to_datetime(order_20['Order Date and Time Stamp'])
  reviews_by_year_month = review.groupby([order_20['Order Date and Time Stamp'].dt.year, order_20['Order Date and Time Stamp'].dt.month])['stars'].count()
  reviews_by_year_month.plot.bar(figsize=(10,6))
  plt.title('reviews of Number of Orders Per Month Per Year')
  plt.xlabel('Number of Orders Per Month Per Year')
  plt.ylabel('reviewes')
  plt.show()
  pdf.savefig(fig)
  # Show the plot
  plt.show()

def analyze_part_of_day():
  fig=plt.figure()
  order_20['Order Date and Time Stamp'] = pd.to_datetime(order_20['Order Date and Time Stamp'])
  order_20['Hour'] = order_20['Order Date and Time Stamp'].dt.hour
  bins=[0, 6, 12, 18, 24]
  labels = ['Night', 'Morning', 'Afternoon', 'Evening']
  order_20['Part_of_Day'] = pd.cut(order_20['Hour'], bins=bins, labels=labels, include_lowest=True)
  order_by_part_of_day = order_20.groupby('Part_of_Day')['Order #'].count()
  order_by_part_of_day.plot.bar(figsize=(10,6))
  plt.title('analysis of Number of Orders Across Parts of a Day')
  plt.xlabel('Number of orders across part of a day')
  plt.ylabel('freq')
  plt.show()
  pdf.savefig(fig)   

def run_all_functions():
  analyze_reviews()
  analyze_payment()
  analyze_state()
  analyze_city()
  analyze_products()
  analyze_r_category()
  analyze_No_order_YM()
  analyze_r_No_order_YM()
  analyze_part_of_day()

choice = input("Enter a number (1-10):")
if choice == '1':
    analyze_reviews()
elif choice == '2':
    analyze_payment()
elif choice == '3':
    analyze_state()
elif choice == '4':
    analyze_city()
elif choice == '5':
    analyze_products()
elif choice=='6':
    analyze_r_category()
elif choice == '7':
    analyze_No_order_YM()
elif choice == '8':
    analyze_r_No_order_YM()   
elif choice == '9':
    analyze_part_of_day()
elif choice=='10':
    run_all_functions()
else:
    print("Invalid choice. Please enter a number between 1 and 10.")
```
