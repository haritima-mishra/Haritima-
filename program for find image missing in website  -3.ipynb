```python
links=[]
product_name=[]
```

```python
for j in range(1, 13):
    url = 'https://yoshops.com/products?page={}'.format(j)
    headers={'User-Agent': 'Mozilla/5.0'}
    response=requests.get(url,headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    product = soup.find_all('div', class_='col-sm-3 col-xs-6')
    for link in product:
        a = link.find('a',class_="product-link")
        b = a.get('href')
        links.append("https://yoshops.com"+b)
        c=link.find('span',class_='product-title').text
        product_name.append(c)       
      
            
```

```python
Description=[]
```

```python
for link in links:
    f = requests.get(link,headers=headers).text
    soup=BeautifulSoup(f,'html.parser')
    d=soup.find('div',class_="single-product-description").text.strip()
    Description.append(d)
```
